```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
default_precision = 0.940610
default_recall = 0.758685
default_F1 = 0.839909
default_MCC = 0.724461
default_BA = 0.855391
```

```python
one_percent_precision = 0.9739591668762432
one_percent_recall = 0.38943942662814934
one_percent_F1 = 0.5564008959882973
one_percent_MCC = 0.47384948967954343
one_percent_BA = 0.6895134749698322
```

```python
two_percent_precision = 0.9678257855822551
two_percent_recall = 0.6126814641459758
two_percent_F1 = 0.7503526724736336
two_percent_MCC = 0.6367317503986472
two_percent_BA = 0.796156799648956
```

```python
five_percent_precision = 0.9375119627999144
five_percent_recall = 0.761207810728782
five_percent_F1 = 0.8402108927625438
five_percent_MCC = 0.7233771687848972
five_percent_BA = 0.8552354920100925
```

```python
comparison = pd.DataFrame({
    "Model":[
    "Default Threshold",
    "One Percent Threshold",
    "Two Percent Threshold",
    "Five Percent Threshold"
    ],
    "Precision":[
    default_precision,
    one_percent_precision,
    two_percent_precision,
    five_percent_precision
    ],
    "Recall":[
    default_recall,
    one_percent_recall,
    two_percent_recall,
    five_percent_recall
    ],
    "F1":[
    default_F1,
    one_percent_F1,
    two_percent_F1,
    five_percent_F1
    ],
    "MCC":[
    default_MCC,
    one_percent_MCC,
    two_percent_MCC,
    five_percent_MCC
    ],
     "BA":[
    default_BA,
    one_percent_BA,
    two_percent_BA,
    five_percent_BA
    ]
}) 
comparison = comparison.round(4)
comparison
```

```python
comparison.to_csv("threshold_comparison.csv", index = False)
```

```python
plt.figure(figsize = (9, 7))
plt.bar(comparison["Model"], comparison["Precision"])
plt.title("Precision Comparison")
plt.ylabel("Precision") 
plt.show()
```

```python
plt.figure(figsize = (9, 7))
plt.bar(comparison["Model"], comparison["Recall"])
plt.title("Recall Comparison")
plt.ylabel("Recall")
plt.show()
```
