```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, matthews_corrcoef, balanced_accuracy_score
```

```python
y_test = pd.read_csv("y_test.csv").squeeze()
y_pred = pd.read_csv("y_pred.csv").squeeze()
```

```python
confusion_mat = confusion_matrix(y_test, y_pred)
```

```python
print("Confusion matrix:\n", confusion_mat)
```

```python
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
bas = balanced_accuracy_score(y_test, y_pred)
```

```python
print("Precision score: ", prec)
print("Recall score: ", rec)
print("F1 score: ", f1)
print("Matthews correlation coefficient: ", mcc)
print("Balanced accuracy score: ", bas)
```

```python
res = pd.DataFrame({
    "Performance metric":["Precision score", "Recall score", "F1 score", "Matthews correlation coefficient", "Balanced accuracy score"],
    "Value":[prec, rec, f1, mcc, bas]
})
res
```
