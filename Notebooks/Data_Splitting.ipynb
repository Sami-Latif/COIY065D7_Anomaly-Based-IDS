```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/sami/Documents/CS Postgrad/IDS/COIY065D7_Anomaly-Based-IDS/Data/cleaned_dataset.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
benign_data = df[df["binary_label"] == 0]
attack_data = df[df["binary_label"] == 1]
```

```python
x_benign_split = benign_data.drop(columns = ["attack_type", "binary_label"])
x_train, x_temp = train_test_split(x_benign_split, test_size = 0.2, random_state = 42)
x_validation, x_test_split = train_test_split(x_temp, test_size = 0.5, random_state = 42)
```

```python
x_attack_split = attack_data.drop(columns = ["attack_type", "binary_label"])
x_attack_split = x_attack_split.sample(n = len(x_test_split), random_state = 42) 
x_test = pd.concat([x_test_split, x_attack_split]) 
```

```python
y_test_benign = [0] * len(x_test_split)
y_test_attack = [1] * len(x_attack_split)
y_test = y_test_benign + y_test_attack
```

```python
print(len(x_test))
print(len(y_test))
```

```python
print("train", x_train.shape)
print("validation", x_validation.shape)
print("test", x_test.shape)
```

```python
x_train.to_csv("x_train.csv", index = False)
x_validation.to_csv("x_validation.csv", index = False)
x_test.to_csv("x_test.csv", index = False)
pd.Series(y_test).to_csv("y_test.csv", index = False)
```
