```python
import os
import pandas as pd
main_folder = "/Users/sami/Documents/CS Postgrad/IDS/COIY065D7_Anomaly-Based-IDS/Data/CSV"
```

```python
entire_data = []
for folder in os.listdir(main_folder): 
    path_folder = os.path.join(main_folder, folder)
    if os.path.isdir(path_folder):
        for sub_folder in os.listdir(path_folder):
            path_sub_folder = os.path.join(path_folder, sub_folder)
            if os.path.isdir(path_sub_folder):
                for file in os.listdir(path_sub_folder):
                    if ".csv" in file:
                        file_path = os.path.join(path_sub_folder, file)
                        print("Loading csv file path: ", file_path)
                        df = pd.read_csv(file_path)
                        df["label"] = sub_folder
                        entire_data.append(df)
print("Total files: ", len(entire_data))
```

```python
df = pd.concat(entire_data, ignore_index = True)
print(df.shape)
```
