```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("/Users/sami/Documents/CS Postgrad/IDS/COIY065D7_Anomaly-Based-IDS/Data/cleaned_dataset.csv")
```

```python
X = df.drop(columns = ["attack_type", "binary_label"])
Y = df["binary_label"]
```

```python
print("X =", X.shape)
print("Y =", Y.shape)
X.head()
```

```python
x_train_KNN, x_temp_KNN, y_train_KNN, y_temp_KNN = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y)
```

```python
x_val_KNN, x_test_KNN, y_val_KNN, y_test_KNN = train_test_split(x_temp_KNN, y_temp_KNN, test_size = 0.5, random_state = 42, stratify = y_temp_KNN)
```

```python
print("KNN Training split:", "x_train_KNN =", x_train_KNN.shape, "y_train_KNN =", y_train_KNN.shape)
print("KNN Validation split:", "x_val_KNN =", x_val_KNN.shape, "y_val_KNN =", y_val_KNN.shape)
print("KNN Testing split:", "x_test_KNN =", x_test_KNN.shape, "y_test_KNN =", y_test_KNN.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_test = pd.read_csv("/Users/sami/Documents/CS Postgrad/IDS/COIY065D7_Anomaly-Based-IDS/Data/x_test.csv")
```

```python
x_train_scale = scaler.fit_transform(x_train_KNN)
x_val_scale = scaler.transform(x_val_KNN)
x_test_scale = scaler.transform(x_test)
```

```python
print(x_train_scale.shape)
print(x_val_scale.shape)
print(x_test_scale.shape)
```

```python
x_train_scale = pd.DataFrame(x_train_scale, columns = x_train_KNN.columns)
print("After scaling:")
x_train_scale[:5]
```
